# Uso de Google Colab con GitHub

[Google Colaboratory](http://colab.research.google.com) es un entorno Jupyter gratuito que se ejecuta completamente en la nube.

Colab permite cargar cuadernos directamente desde GitHub y guardar cuadernos nuevamente en GitHub.

## Cargar cuadernos públicos directamente desde GitHub

Colab puede cargar cuadernos públicos alojados en GitHub directamente, sin necesidad de realizar ningún paso de autorización.

Por ejemplo:

https://github.com/googlecolab/colabtools/blob/main/notebooks/colab-github-demo.ipynb

Enlace directo en Colab:

https://colab.research.google.com/github/googlecolab/colabtools/blob/main/notebooks/colab-github-demo.ipynb

### Crear un enlace "Abrir en Colab"

Recuerde reemplazar la URL del cuaderno por la suya.

## Guardar una copia en GitHub

1. Iniciar sesión en Google.
2. Abrir el cuaderno desde GitHub en Colab.
3. Archivo → Guardar una copia en GitHub.
4. Elegir repositorio, rama y nombre.
