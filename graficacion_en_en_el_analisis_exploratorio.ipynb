```python
pinguinos = sns.load_dataset("penguins")
```

```python
pinguinos['body_mass_g'].hist()
plt.title('Distribucin de indice de masa de pinguinos')
plt.xlabel('Masa corporal(g)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.pairplot(data=pinguinos,hue = 'species')
plt.show()
```

```python
sns.scatterplot(data=pinguinos, x= 'flipper_length_mm', y = 'body_mass_g',hue='species',hue_order=['Gentoo','Adelie','Chinstrap'])
plt.show()
```

```python
g = sns.scatterplot(data=pinguinos, x= 'flipper_length_mm', y = 'body_mass_g',hue='species',hue_order=['Gentoo','Adelie','Chinstrap'])
```

```python
print(type(g))
```

```python
sns.relplot(data=pinguinos, x= 'flipper_length_mm', y = 'body_mass_g', kind= 'scatter', hue='species', col = 'island',row= 'sex')
plt.show()
```

```python
sns.relplot(data=pinguinos, x= 'flipper_length_mm', y = 'body_mass_g', kind= 'line', hue='species', col = 'island',row= 'sex')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(data=pinguinos, x = 'species', hue= 'species', palette='colorblind')
```

```python
sns.scatterplot(data=pinguinos, x = 'flipper_length_mm' , y='bill_depth_mm', hue='species',size='body_mass_g' )
plt.title('Grafico 1')
plt.show()
```

```python
pinguinos['species'].value_counts(normalize=True).plot(kind='bar')
plt.show()
```

```python
sns.barplot(data= pinguinos, x = 'species', y = 'bill_length_mm')
```

```python
sns.countplot(data=pinguinos, x='species')
plt.show()
```

```python
sns.boxplot(data=pinguinos, x='species', y = 'bill_length_mm', hue= 'sex', palette='husl', hue_order=['Female', 'Male'])
plt.show()
```
